```python
import weighting
import pandas as pd
import networkx as nx
```

```python
edge = weighting.G
edge
```

```python
# jumlah node dan sisi

nodes = nx.number_of_nodes(edge)
edges = nx.number_of_edges(edge)

print('Jaringan ini terdiri dari', nodes, 'node dan', edges, 'edges.')
```

```python
# returns the number of selfloop edges

selfloop = nx.number_of_selfloops(edge)
print('Pada jaringan ini terdapat', selfloop)
```

```python
#remove selfloop
edge.remove_edges_from(nx.selfloop_edges(edge))
print('Selfloop berhasil dihapus!')
```

```python
# create edge list
edge_list = nx.to_pandas_edgelist(edge)
edge_list
```

```python
edge_list.to_csv('edge_list.txt', sep=' ', index=False, header=False)
```
